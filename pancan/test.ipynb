```python
fu = pd.read_csv('clinical_PANCAN_patient_with_followup.tsv',sep='\t', encoding='latin1')
nvalues = []
for i in fu.columns: 
    fu_drop = fu[i].dropna()
    fu_drop = fu_drop[fu_drop!='[Not Available]']
    fu_drop = fu_drop[fu_drop!='[Not Applicable]']
    nvalues.append([i,len(fu_drop)])
nvalues = pd.DataFrame(nvalues,columns=['column','nvalues'])
nvalues_sorted = nvalues.sort_values('nvalues',ascending=False)
for i in nvalues_sorted.iterrows():
    print(i[1]['column'],'\t\t\t',i[1]['nvalues'])
```
