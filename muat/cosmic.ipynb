```python
import os
import shutil

# Path to the directory containing the files
parent_directory = '/data/dandreas/SomaticMutationsLLM/vcf_for_cosmic_test/'
directory = parent_directory+f'143_vcf/'
# Iterate over all files in the directory
for filename in os.listdir(directory):
    filepath = os.path.join(directory, filename)
    if os.path.isfile(filepath):
        # Create a new directory with the name of the file (without the extension)
        new_dir = os.path.join(f'{parent_directory}143_folders/', filename.split('.')[0])
        os.makedirs(new_dir, exist_ok=True)
        # Copy the file into the new directory
        shutil.copy(filepath, new_dir)
```
